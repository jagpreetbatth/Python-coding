# Create Loyalty Flag

```python
# Create column with each user_id's max order

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Code to remove max on rows displayed

pd.options.display.max_rows = None
```

```python
# Verify Column Creation

ords_prods_merge.head(100)
```

```python
# Derive columns with loc()

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check Results

ords_prods_merge[['user_id', 'loyalty_flag', 'max_order']].head(60)
```

```python
# Use loyalty flag to check basic statistics of product prices for each loyalty category

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### Note the mean price of New Customers is higher than that of the regular and loyal customers, this can be due to loyality coupons, rewards or algorithms that automatically decrease prices for those who have been with the company for a longer duration of time.

```python
# Derive column with each user_id's mean purchase price

ords_prods_merge['average_purchase_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Verify Column Creation

ords_prods_merge[['user_id', 'average_purchase_price', 'loyalty_flag']].head()
```

```python
# Derive new flag column using loc() function based on high and low spenders

ords_prods_merge.loc[ords_prods_merge['average_purchase_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['average_purchase_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check Results

ords_prods_merge[['user_id', 'average_purchase_price', 'spending_flag']].head(100)
```

```python
# Derive column with each user_id's median days_since_prior_order

ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Verify Column Creation

ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_since_prior_order']].head()
```

```python
# Derive new flag columns with loc() based on frequent, non-frequent and Regular customers

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Check Results

ords_prods_merge[['user_id', 'median_days_since_prior_order', 'order_frequency_flag']].head(100)
```
